```python
import pandas as pd
from iminuit import Minuit 
from iminuit.cost import LeastSquares
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
def graph(X ,Y, Z):
    K = np.zeros(len(X))
    plt.plot(X, Y, label='data')
    plt.plot(X, Z, label='fit')
    plt.plot(X, K)
    plt.show()
```

```python
def read_remove(file_path):
    """
    Read data from a CSV file and return the bins and counts.
    """
    Z = pd.read_csv(file_path, sep='\t+', header=None, engine='python')
    return(Z)
```

```python
B = read_remove('Bkg_55.5V.txt')
C = read_remove('Cobalto_55.5V.txt')
S = read_remove('Sodio_55.5V.txt')

C_puro = C[1] - B[1]
S_puro = S[1] - B[1]

print(C_puro)
print(S_puro)
```

```python
graph(B[0], C[1], C_puro)
graph(B[0], S[1], S_puro)
```
